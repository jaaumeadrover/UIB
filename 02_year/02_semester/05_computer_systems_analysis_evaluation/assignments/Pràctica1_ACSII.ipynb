**1. ¿Qué servidor resulta más adecuado para el cambio sólo considerando el rendimiento? ¿Por qué? ¿En qué métrica o valor determina la decisión?**

<blockquote>Para decidir que servidor es más adecuado para nuestro cambio, determinaremos el rendimiento de una de las maneras estudiadas en clase. El rendimiento se puede obtener mediante la fórmula del cambio relativo:</blockquote>

$$ CambioRelativo=\frac{T_b-T_a}{T_b}=\frac{26.998-23,584}{23,584}>0 $$

En consecuencia, podemos afirmar que si el resultado es positivo, el servidor B es más lento que A, condición que se cumple.
El valor resultante es el que nos ayuda a determinar la decisión.


 **2. ¿Y si además tenemos en cuenta el coste del servidor? ¿Cuál sería más adecuado? ¿Por qué? ¿En qué métrica valor te basas?**
 
 <blockquote>Si tenemos en cuenta el coste del servidor, realizaremos otro cálculo de la medida rendimiento/coste y compararemos ambos valores.Estos nos indican la relación que hay entre el rendimiento y el coste de un componente. El servidor más adecuado será el que tenga dicho valor más grande respecto al otro. Veamos los siguientes cálculos</blockquote>


$$ \frac{Coste_A}{Coste_B}=\frac{1245}{907}=1.37\newline $$
<blockquote>El servidor A es 1.37 veces más caro o un 37% más caro que el servidor B.</blockquote>

<div align="center"><b> Servidor A </b></div><br>
$$ \frac{Rendimiento_A}{Coste_A}=\frac{1}{Tiempo_A · Coste_A}=\frac{1}{23.584 · 1245}=3.4x10^{-5}\newline $$
<br><br>
<div align="center"><b> Servidor B </b></div><br>
$$ \frac{Rendimiento_B}{Coste_B}=\frac{1}{Tiempo_B · Coste_B}=\frac{1}{26.998 · 907}=4.08x10^{-5}\newline $$

<blockquote>Como podemos observar,la relación rendimiento/coste es mayor en el servidor B, por lo que podemos afirmar que este es el más adecuado. Otro dato que respalda nuestra elección es que el servidor A es un 37% más caro que el servidor B. Me baso principalmente en el valor del rendimiento/coste, métrica valor elegida.</blockquote>
